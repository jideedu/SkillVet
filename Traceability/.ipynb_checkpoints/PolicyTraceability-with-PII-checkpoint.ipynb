# Getting the Skill Description and Policy Link from the Amazon US Amazon

```python
import collections
import pandas as pd

models = PolicyAnalyser.LoadModels()

#'Load the CSV'
df = pd.read_csv('Skill_with_PII.csv')
df['SKILL']= df['SKILL'].str.strip().str.lower()
df['DEVELOPER']= df['DEVELOPER'].str.strip().str.lower()
df['INFO']= df['INFO'].str.strip().str.lower()

df['Permission_from_Api'] = ''
df['Combine_Permission'] = ''
df['Privacy_Policy_link'] = ''
df['Description'] = ''
count = 0

seen_skill = []

with open('file_US.json', 'r', encoding='utf8')as fp:
    for i in fp:
        jsonobj = json.loads(i)
        try:
            permission = []
            name = (jsonobj['Name']).strip().lower()         
            dev = (jsonobj['Developer']).strip().lower()
            if name+dev not in seen_skill:
                seen_skill.append(name+dev)
            
                for index, row in df.iterrows():
                    if (row['SKILL']== name) and (row['DEVELOPER']== dev):
                        count = count + 1
                        b = row['INFO'].split(',')
                        for elem in b:
                            permission.append(elem)
                        if jsonobj['Skill_permission'] !=[]:
                            row['Permission_from_Api'] = jsonobj['Skill_permission']
                            for perm in jsonobj['Skill_permission']:
                                perm = perm.strip().lower()
                                if perm not in permission:
                                    permission.append(perm)              

                        if len (permission) > 0:
                            #print(count, ':', name, ':', permission)
                            row['Combine_Permission'] = permission
                            row['Description'] = (jsonobj['Skill_description']).replace('Description\n','')
                            if jsonobj['privacy_policy']:
                                row['Privacy_Policy_link'] = jsonobj['privacy_policy']
                            else:
                                row['Privacy_Policy'] = ''
                        
        except:
            pass
```

# Getting Policy Online to the csv file

```python
from selenium.webdriver.firefox.options import Options
from selenium import webdriver
from bs4 import BeautifulSoup as bs

# Create a new instance of the Mozilla driver
options = Options()
options.headless = True
driver = webdriver.Firefox(options=options, executable_path="geckodriver.exe")


df = pd.read_csv('Details_P.csv')
df['Policy']=''


def tag_visible(element):
    if element.parent.name in ['script', 'style', 'head', 'title', 'meta', '[document]']:
        return False
    if isinstance(element, Tag):
        return False
    if isinstance(element, Comment):
        return False
    return True

def text_from_html(body):
    soup = BeautifulSoup(body, "lxml")
    texts = soup.findAll(text=True)
    visible_texts = filter(tag_visible, texts)  
    return u" ".join(t.strip() for t in visible_texts)

for index, row in df.iterrows():
    try:
        url =  row['Privacy_Policy']
        driver.get(url.strip())
        page = driver.page_source.lower()
        cleancode = text_from_html(page)
#         print(cleancode)
        row['Policy'] = cleancode
    except:
        print(row['SKILL'])
        pass


df.to_csv('Details_Pii_new.csv', index=None)
```

```python
import PolicyAnalyser
from os import listdir
from os.path import isfile, join
import random
import collections

models = PolicyAnalyser.LoadModels()

#policiespath = '/Users/admin/Downloads/New_Crawl_Data_policies/AU'
policiespath = 'data/Policy_from_Converse/'


files = [join(policiespath, f) for f in listdir(policiespath) if isfile(join(policiespath, f))]

result = []
olderperm = [('Lists Read Access', 'Personal Information'),
             ('Lists Write Access', 'Personal Information'),
             ('lists write access (2)', 'Personal Information'),
             ('First Name', 'Name'),
             ('Full Name', 'Name'),
             ('Given Name', 'Name'),
             ('Device Country and Postcode', 'device country and postal code'),
             ('phone number','Mobile Number'),
             ('birthday', 'Personal Information'),
             ('zip', 'device country and postal code'),
             ('address','Device Address'),
             ('location','Location Services'),
             ('email','Email Address'),
             ('area code', 'device country and postal code'),
             ('gender', 'Personal Information'),
             ('born','Personal Information'),
             ('zipcode', 'device country and postal code'),
             ('postal code', 'device country and postal code') 
            ]
olderrem = ['Alexa Notifications', 'Skill Personisation', 'Reminders', 'Timer']

#sample_policies = random.sample(files, 500)
sample_policies = files
processedpolicies = []
ignoredpolicies = 0

for policyfile in sample_policies:
    
    print()
    print('Processing policy', policyfile);
    try:
        
        splitname = policyfile.split('__')
        #splitname = policyfile.split('..')
        skillname = splitname[0][splitname[0].rfind('/')+1:].strip()
        print('\tSkillname:', skillname)
        dev = splitname[1].strip() 
        print('\tDev:', dev)
        permissions_requested = policyfile.split('[')[1].replace(']', '').replace('.txt', '').replace('\'', '').split(',')
        permissions_requested = [k.strip() for k in permissions_requested]
        print('\tPermissions requested', permissions_requested)
        
        #check if its a repeated policy
        if(skillname+'__'+dev in processedpolicies):
            continue
        processedpolicies.append(skillname+'__'+dev)
        
        #adapt older permissions to new ones
        for perm in olderperm:
            permissions_requested = [perm[1] if x==perm[0] else x for x in permissions_requested]
        #ignore non existing actual permissions
        permissions_requested = [x.lower()  for x in permissions_requested if x not in olderrem and len(x)>0]
        print('\tAfter adapting permissions', permissions_requested)
        
        #if the file does not request any permission, we ignore it and jump to next
        #privacy policy
        if(len(permissions_requested) == 0):
            to['ignored'] += 1
            tf['ignored'] += 1
            continue
                 
        perm_found = PolicyAnalyser.AnalysePolicy(
            policyfile, 
            models, 
            smartNegation = True, 
            smartNone = True, 
            kwfilter = False, 
            kw_ignore_filter=True,
            splitintolines = True)
        permissions_found = [p.lower() for p in perm_found[0]]        
        
        
        #if we find 'device address' we also find 'device country and postal code', because the latest is a subset
        #of the first.
        if('device address' in permissions_found):
            permissions_found.append('device country and postal code')
        print('\tPermissions found in policy', permissions_found)


        #check whether permissions requested are a subset of the permissions found - complete
        #if no gt is found, we do not perform the traceabilaity analysis
        #replace personal information for any permission found in the policy
        #if there is personal information in permissoins requested, and we have some permission founds that are not
        #none, we replace personal information for the first permission requested
        permissions_found_not_none = [x for x in permissions_found if x != 'none']
        if('personal information' in permissions_requested and len(permissions_found_not_none)>0):
           #find non non permission found
           permissions_requested = [permissions_found_not_none[0] if x=='personal information' 
                                    else x for x in permissions_requested]

        typetraceability = None
        if( set(permissions_requested).issubset(set(permissions_found)) ):
            typetraceability = 'c'
        elif( (len(permissions_found) == 0 or (len(permissions_found) == 1 and 'none' in permissions_found)) 
             and len(permissions_requested)>0):
            typetraceability = 'b'
        else:
            #check if any permission found fits with the permission requested by the skill
            typetraceability = 'b'
            for p in permissions_requested:
                if(p in permissions_found or 'personal information' in permissions_found):
                    typetraceability = 'p'
                    break            
        
        result.append(typetraceability)        
        print('\tComputed Traceability', typetraceability)             
    except Exception as e:
        print(e)
    
    
print (collections.Counter(result))

```
